```python
d = dict()

def is_word(w):
    return w.isalpha()

for i in range(2000):
    s = df.iloc[i]['clean_text']
    s = eval(s)
    for j, w in enumerate(s):
        if not is_word(w['Token']):
            continue
        w['Token'] = w['Token'].lower()
        if w['Token'] not in d:
            d[w['Token']] = []
        d[w['Token']].append([w['POS'], i, j])

for tok in d:
    dd = dict()
    for pos, i, j in d[tok]:
        if pos not in dd:
            dd[pos] = []
        dd[pos].append((i, j))
    if len(dd) > 1:
        print(tok, d)
        break

# видимо, pymorph не различает слова в разных контекстах
```
